```python
import pandas as pd
from sklearn.cluster import KMeans
```

```python
dataset = pd.read_csv('iris.csv')
classes = dataset['variety']
dataset = dataset.drop('variety', axis=1)
kmeans = KMeans(n_clusters=3, max_iter=300)
kmeans.fit(dataset)
dataset
```

```python
print(kmeans.labels_)
```

```python
import matplotlib.pyplot as plt

colors = ['g.', 'r.', 'b.']
for i in range(len(dataset)):
  plt.plot(dataset.iloc[i]['sepal.length'],
           dataset.iloc[i]['sepal.width'],
           colors[kmeans.labels_[i]],
           markersize=10)
plt.show();
```

```python
#Convertendo Texto para Inteiro
classes[classes == 'Setosa'] = 0
classes[classes == 'Versicolor'] = 1
classes[classes == 'Virginica'] = 2
classes
```

```python
from sklearn.metrics import classification_report
print(classification_report
     (classes.values.astype(int),
      kmeans.labels_))
```

```python
# Como encontrar a quantidade de clusters ideal

wcss = []

for i in range(1, 11):
  kmeans = KMeans(n_clusters=i, max_iter=300)
  kmeans.fit(dataset)
  wcss.append(kmeans.inertia_)
  print(kmeans.inertia_)

plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show();
```
